# 等值连接

```python
!mysql -e "use nba;SELECT player_id, player.team_id, player_name, height, team_name FROM player, team WHERE player.team_id = team.team_id"
```

# 非等值连接

```python
#身高等级
!mysql -e "use nba;SELECT p.player_name, p.height, h.height_level FROM player AS p, height_grades AS h WHERE p.height BETWEEN h.height_lowest AND h.height_highest"
```

# 外连接

```python
# left join +表示从表
!mysql -e "use nba;SELECT * FROM player left join team where player.team_id = team.team_id"
```

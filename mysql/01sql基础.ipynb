```python
#空字段
!mysql -e"use heros_data;SELECT name,mp_max, hp_max FROM heros where hp_max is NULL;"
```

```python
#最大生命值加最大法力值大于 8000 的英雄，或者最大生命值大于 6000 并且最大法力值大于 1700 的英雄
!mysql -e"use heros_data;SELECT name,mp_max, hp_max FROM heros where (mp_max+hp_max)>8000 or (hp_max >6000 and mp_max >1700) ;"
```

```python
#主要或次要角色为法师或是射手，英雄上线时间不在2016-01-01和2017-01-01之间
!mysql -e"use heros_data;SELECT * FROM heros where role_main in ('法师','射手') or role_assist in  ('法师','射手') and DATE(birthdate) not between '2016-01-01' and  '2017-01-01' order by (hp_max +mp_max) desc;"
```

```python
#16年10月1日后上线的英雄平均最大生命值，平均最大法攻值，最大物防
!mysql -e"use heros_data;select avg(hp_max),avg(mp_max),max(attack_max) FROM heros WHERE DATE(birthdate)>'2016-10-01';"
```

筛选最大生命值大于 6000 的英雄，按照主要定位进行分组，选择分组英雄数量大于 5 的分组，按照分组英雄数从高到低进行排序，并显示每个分组的英雄数量、主要定位和平均最大生命值。

```python
!mysql -e"use heros_data;select count(*) as num,role_main,avg(hp_max) from heros where hp_max >6000 group by role_main having num>5 order by num desc;" 
```

```python
!mysql -e"use heros_data;select count(*) as num,avg(hp_max+mp_max),max(hp_max+mp_max) from heros where hp_max >6000 group by role_main having num>5 order by num desc;" 
```
